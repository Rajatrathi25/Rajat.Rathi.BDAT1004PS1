# Ques 3) 
Implement function triangleArea(a,b,c) that takes as input the lengths of the 3
sides of a triangle and returns the area of the triangle. By Heron's formula, the area
of a triangle with side lengths a, b, and c is
s(s - a)(s -b)(s -c)
, where
s = (a+b+c)/2.
> triangleArea(2,2,2)

1.7320508075688772


```
class triangleArea{
    
    public int Area(double a,double b,double c)
    {
        double s = (a+b+c)/2;
        double root = s*(s-a)*(s-b)*(s-c);
        //Math.sqrt is used to sqrt the value
        double AreaTriangle = Math.Sqrt(root);
        Console.WriteLine("Area of Triangle is {0}",AreaTriangle);
        return 0;
    }
    
}
//object is created of the function
triangleArea r = new triangleArea();
//call is made to the function
r.Area(2,2,2);
```

Write a program in C# Sharp to separate odd and even integers in separate arrays. 

```
// This question has been solved by me in Visual studio as I was not able to figure out how to give input here in jupter and I have attached an image of the output.
Console.WriteLine("Hey,I will be finding Even and odd integers");

int noOfValues;
Console.WriteLine("Enter the number of Values");
noOfValues = Convert.ToInt32(Console.ReadLine());
/* Converts to integer type */

/* Takes the input of values from the user*/
int[] array = new int[noOfValues];
String inputVal;
for (int i = 0; i < noOfValues; i++) { 
    Console.WriteLine("Enter the {0}",i);
    inputVal=Console.ReadLine();
    array[i] = Convert.ToInt32(inputVal);
}

// Seperate out the even and odd values
var evenList = new List<int>();
var oddList = new List<int>();

for (int i = 0; i < array.Length; i++) {
    if (array[i] % 2 == 0)
    {
        evenList.Add(array[i]);
    }
    else { 
        oddList.Add(array[i]);  
    }
}

// prints the even values
Console.WriteLine("Even Integers are :");
for (int i = 0; i < evenList.Count; i++) {
    Console.WriteLine(evenList[i]);
}

//prints the odd values 
Console.WriteLine("Odd Integers are:");
for (int i = 0; i < oddList.Count; i++) { 
    Console.WriteLine(oddList[i]);
}
```

depending on whether the point (x,y) lies in the rectangle with lower left

b)  Use function inside() from part a. to write an expression that tests whether
the point (1,1) lies in both of the following rectangles: one with lower left
corner (0.3, 0.5) and upper right corner (1.1, 0.7) and the other with lower
left corner (0.5, 0.2) and upper right corner (1.1, 2). 

```
// The class is created
class Rectangle{

    //function for a part to find the points inside a rectangle
    public bool pointlies(int x,int y,int x1,int y1,int x2,int y2){
        
        // condition to check whether it lies in the rectangle 
        if(x>x1 && y>y1 && x<x2 && y<y2){
            return true;
        }
        else{
            return false;
        }
    }
    // function for b part to find the points in two rectangles 
    public bool inside(double x , double y, double x1T1,double y1T1,double x2T1,double y2T1 ,double x1T2,double y1T2,double x2T2,double y2T2 ){
        
        // condition to check whether it lies in both the rectangles
        // this if condition checks for first rectangle
        if(x>x1T1 && y>y1T1 && x<x2T1 && y<y2T1){
            // this if condition checks for second rectangle
            if(x>x1T2 && y>y1T2 && x<x2T2 && y<y2T2){
                return true;
            }
            // this else condition state the if it is not present in second rectangle return false
            else{
                return false;
            }
            
        }
        // this else condition states that if initially its not present in first rectangle than return false 
        else{
            return false;
        }
        
    }
}

// obj is created and calls are made for a part
Rectangle newRectangle = new Rectangle();
bool Val=newRectangle.pointlies(1,1,0,0,2,3);
Console.WriteLine(Val);

// call is made for second part
bool insideVal = newRectangle.inside(1,1,0.3,0.5,1.1,0.7,0.5,0.2,1.1,2);
Console.WriteLine(insideVal);
```
